```python
import pandas as pd
import os
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\arya\projects\covid_watch\data\final\results_20210110.csv")
```

## Confirmed Cases

```python
df_confirmed = df[df['status']=="Confirmed"].groupby(['ds','country']).agg({'yhat':sum}).reset_index(level=['ds', 'country'])
fig = px.line(df_confirmed, x="ds", y="yhat", color='country')
fig.show()
```

## Deaths

```python
df_deaths = df[df['status']=="Deaths"].groupby(['ds','country']).agg({'yhat':sum}).reset_index(level=['ds', 'country'])
fig = px.line(df_deaths, x="ds", y="yhat", color='country')
fig.show()
```

## Recovered

```python
df_recovered = df[df['status']=="Recovered"].groupby(['ds','country']).agg({'yhat':sum}).reset_index(level=['ds', 'country'])
fig = px.line(df_recovered, x="ds", y="yhat", color='country')
fig.show()
```

## Death Rate

```python
df_death_rate = pd.merge(left=df_deaths, right=df_confirmed, on=["ds", "country"], how="left")
df_death_rate["death_rate"] = df_death_rate["yhat_x"]/df_death_rate["yhat_y"]
fig = px.line(df_death_rate, x='ds', y='death_rate', color='country')
fig.show()
```

## Recovery Rate

```python
df_recovery_rate = pd.merge(left=df_recovered, right=df_confirmed, on=["ds", "country"], how="left")
df_recovery_rate["recovery_rate"] = df_recovery_rate["yhat_x"]/df_recovery_rate["yhat_y"]
fig = px.line(df_recovery_rate, x='ds', y='recovery_rate', color='country')
fig.show()
```
